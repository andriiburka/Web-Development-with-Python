3.List Statistics
==
---------------

You will be given a number n. On the next n lines you will receive
integers. You have to create and print two lists:

-   One with all the positives (including 0)

-   One with all the negatives

Finally print the following message**: "Count of positives:
{count\_positives}. Sum of negatives: {sum\_of\_negatives}"**

### Example

<table>
<thead>
<tr class="header">
<th><strong>Input</strong></th>
<th><strong>Output</strong></th>
</tr>
</thead>
<tbody>
<tr class="odd">

<td><p>5<br>10<br>3<br>2<br>-15<br>-4</p></td>

<td><p><strong>[10, 3, 2]</strong></p>
<p><strong>[-15, -4]</strong></p>
<p><strong>Count of positives: 3. Sum of negatives: -19</strong></p></td>
</tr>
</tbody>
</table>

>### Hints
- We start by reading the number n:
- Then, we create **a loop**, we read the **current number** and check if it **is greater than zero or not**
- If it is, we add it to the positives list
- If it is not, we add it to the negatives list
- Then we print the three lines:
- To get the count of the positives we can use the **len** function
- To get the sum of the negatives we can use the **sum** function

```python
n = int(input())
list_positive = []
list_negative = []

for num in range(1, n + 1):
    number = int(input())
    if number < 0:
        list_negative.append(number)
    else:
        list_positive.append(number)

print(list_positive)
print(list_negative)
print(f'Count of positives: {len(list_positive)}. Sum of negatives: {sum(list_negative)}')
```
